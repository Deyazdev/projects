```python
#clean up data
# We only want 50k or 50 mi races in the USA
marathon.head()
```

```python
marathon_USA = marathon.loc[(marathon['Event distance/length'].isin(['50km', '50mi'])) & (marathon['Event name'].str.contains(r'\(USA\)'))]
marathon_USA
```

```python
# add event country to data
marathon['Event Country'] = marathon['Event name'].str.extract(r'\(([A-Za-z]+)\)')
```

```python
#Check if USA has any other coding in the name
invalid_countries = marathon[~marathon['Event Country'].str.contains(r'^[A-Z]{3}$', na=False)]
invalid_countries['Event Country'].unique()
```

```python
#Narrow down to races in 2020
marathon_USA = marathon_USA[marathon_USA['Year of event'] == 2020 ]
marathon_USA.head()
```

```python
marathon_USA['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
#We will remove USA from event name as we have a column for the country
marathon_USA['Event name'] = marathon_USA['Event name'].str.split('(').str.get(0)
marathon_USA.head()
```

```python
#create column for athlete age
marathon_USA['Athlete age'] = marathon_USA['Year of event'] - marathon_USA['Athlete year of birth']
marathon_USA['Athlete age'] = marathon_USA['Athlete age'].fillna(0).astype(int)
marathon_USA['Athlete age'].describe()
```

```python
marathon_USA[ marathon_USA['Athlete age'] == 0]
```

```python
marathon_USA['Athlete performance'] = marathon_USA['Athlete performance'].str.split(' ').str.get(0)
```

```python
# let us now drop columns: athlete club, athlete age category, athlete year of birth, athlete age category
marathon_USA = marathon_USA.drop(['Athlete club',
       'Athlete country',
       'Athlete age category'], axis = 1)
```

```python
marathon_USA['Athlete year of birth'] = pd.to_datetime(marathon_USA['Athlete year of birth'], format='%Y')
```

```python

marathon_USA
```

```python
marathon_USA.isna().sum()
```

```python
marathon_USA = marathon_USA.dropna()
marathon_USA.head()
```

```python
marathon_USA.duplicated(keep= False).sum()
```

```python
marathon_USA.reset_index(drop = True, inplace=True)
```

```python

marathon_USA['Athlete average speed'] = marathon_USA['Athlete average speed'].astype(float)
marathon_USA['Athlete performance'] = pd.to_timedelta(marathon_USA['Athlete performance'])
marathon_USA.dtypes
```

```python
marathon_USA.rename(columns= {
'Year of event':'year of event',
'Event dates':'event dates',
'Event name':'event name',
'Event distance/length':'event distance/length',
'Event number of finishers':'event number of finishers',
'Athlete performance':'athlete performance',
'Athlete year of birth':'athlete year of birth',
'Athlete average speed':'athlete average speed',
'Athlete ID':'athlete id',
'Athlete age':'athlete age'
},inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=marathon_USA, x = marathon_USA['Athlete gender'], y = marathon_USA['Athlete gender'].count())
```

```python
sns.histplot(marathon_USA, x = 'event distance/length', hue = 'Athlete gender')
```

```python
sns.displot(marathon_USA[marathon_USA['event distance/length'] == '50mi']['athlete average speed'])
```

```python

sns.lmplot(marathon_USA, x = 'athlete age', y = 'athlete average speed', hue= 'Athlete gender', height= 5, aspect=3)
```

```python
# difference in speed between male and female racers in 50k races
sns.barplot(marathon_USA, x = 'Athlete gender', y  = 'athlete average speed', hue = 'event distance/length')
```

```python
# what ages are the best to run in a 50mi race? 
best_ages = marathon_USA[marathon_USA['event distance/length'] == '50mi'].groupby('athlete age')['athlete average speed'].agg(['mean', 'count']).query('count>20').sort_values(by = 'mean', ascending = False)
sns.lineplot(best_ages, x = 'athlete age', y = 'mean')
```

```python
# average speed in different seasons? 
marathon_USA['month of race'] = marathon_USA['event dates'].str.split('.').str.get(1).astype(int)
marathon_USA.head()
```

```python
marathon_USA['month of race'] = marathon_USA['month of race'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter'  )
```

```python
#checking to see if it worked
marathon_USA['month of race'].value_counts()
```

```python
marathon_USA[marathon_USA['event distance/length'] == '50mi'].groupby('month of race')['athlete average speed'].agg(['mean', 'count']).reset_index().sort_values(by = 'mean', ascending = False)
```
